```python
import pandas as pd

file_path = r"C:\Users\Kasper Hassing\Desktop\Speciale_KryptoSentiment\data\market_data\BTCUSDT_features.csv"
df = pd.read_csv(file_path)

# 1. Tjek for NaN-værdier før behandling
print("NaN-værdier før behandling:")
print(df.isna().sum())

# 2. Konverter og sorter datoer
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date')

# 3. Filtrer på de ønskede datoer 
start_date = pd.to_datetime('2021-01-01')
end_date = pd.to_datetime('2022-02-28')
df = df[(df['date'] >= start_date) & (df['date'] <= end_date)]

# 4. Tjek for NaN-værdier efter behandling
print("\nNaN-værdier efter behandling:")
print(df.isna().sum())

# 5. Gem den nye fil
new_file_path = file_path.replace('.csv', '_cleaned.csv')
df.to_csv(new_file_path, index=False)

print(f"\nNy fil gemt som: {new_file_path}")
print(f"Antal rækker før: {len(pd.read_csv(file_path))}, efter: {len(df)}")
print(f"Første dato: {df['date'].min()}, Sidste dato: {df['date'].max()}")
```
