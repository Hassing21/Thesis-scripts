```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# === DATA: 7-day forecast results ===
data = {
    'Model': ['XGBoost', 'XGBoost', 'LSTM', 'LSTM', 'TFT', 'TFT'],
    'Window': ['W1', 'W2'] * 3,
    'Correlation': [-0.1493, 0.0585, -0.1558, 0.0047, 0.6537, 0.6816],
    'Hit Rate': [0.4096, 0.4691, 0.6415, 0.5882, 0.6087, 0.5714]
}
df = pd.DataFrame(data)

# === STYLE SETTINGS ===
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("colorblind")
sns.set_context("notebook", font_scale=1.2)
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['figure.figsize'] = (12, 8)

# === Correlation Plot ===
plt.figure()
ax1 = sns.barplot(data=df, x='Model', y='Correlation', hue='Window')
plt.title('Correlation  Forecasting future_return_7d', fontsize=14, fontweight='bold')
plt.ylim(-0.2, 0.75)
plt.ylabel('Correlation', fontsize=12, fontweight='bold')
plt.xlabel('', fontsize=12, fontweight='bold')
plt.legend(title='Window')

# Add value labels
for p in ax1.patches:
    height = p.get_height()
    if height != 0.0:
        if height > 0:
            ax1.annotate(f'{height:.3f}',
                         (p.get_x() + p.get_width() / 2, height + 0.005),
                         ha='center', va='bottom', fontsize=11, fontweight='bold', color='black')
        else:
            ax1.annotate(f'{height:.3f}',
                         (p.get_x() + p.get_width() / 2, height - 0.005),
                         ha='center', va='top', fontsize=11, fontweight='bold', color='black')

plt.tight_layout()
plt.show()

# === Hit Rate Plot ===
plt.figure()
ax2 = sns.barplot(data=df, x='Model', y='Hit Rate', hue='Window')
plt.title('Hit Rate  Forecasting future_return_7d', fontsize=14, fontweight='bold')
plt.ylim(0.4, 0.7)
plt.ylabel('Hit Rate', fontsize=12, fontweight='bold')
plt.xlabel('', fontsize=12, fontweight='bold')
plt.legend(title='Window')

# Add value labels
for p in ax2.patches:
    height = p.get_height()
    ax2.annotate(f'{height:.3f}',
                 (p.get_x() + p.get_width() / 2, height + 0.005),
                 ha='center', va='bottom', fontsize=11, fontweight='bold', color='black')

plt.tight_layout()
plt.show()
```
