```python
import pandas as pd
from tqdm import tqdm

path = r"C:\Users\Kasper Hassing\Desktop\Speciale_KryptoSentiment\data\twitter_posts\bitcoin_tweets_cleaned_07.csv"
output_path = r"C:\Users\Kasper Hassing\Desktop\Speciale_KryptoSentiment\data\twitter_posts\bitcoin_tweets_cleaned_08.csv"

# === IndlÃ¦s data ===
print("ğŸ“¥ IndlÃ¦ser tweets...")
df = pd.read_csv(path)

# === Beregn antal ord pr. tweet ===
print("ğŸ”¢ Beregner ordlÃ¦ngde...")
df['text_length'] = df['text'].apply(lambda x: len(str(x).split()))

# === Statistik fÃ¸r filtrering ===
print("\nğŸ“Š Statistik fÃ¸r filtrering:")
original_count = len(df)
print(f"Total antal tweets fÃ¸r filtrering: {original_count:,}")

# === Fjern tweets med mere end 40 ord ===
print("\nğŸ§¹ Fjerner tweets med mere end 40 ord...")
tqdm.pandas(desc="Filtrerer")
df = df[df['text_length'] <= 40]

# === Statistik efter filtrering ===
filtered_count = len(df)
removed_count = original_count - filtered_count
removed_pct = (removed_count / original_count) * 100

print("\nğŸ“Š Statistik efter filtrering:")
print(f"Tweets efter filtrering: {filtered_count:,}")
print(f"Fjernede tweets: {removed_count:,} ({removed_pct:.2f}%)")

# === Gem det filtrerede datasÃ¦t ===
print(f"\nğŸ’¾ Gemmer det filtrerede datasÃ¦t til:\n{output_path}")
df.to_csv(output_path, index=False)

print("\nâœ… Filtrering fuldfÃ¸rt!")
```

```python
import pandas as pd
import re
from tqdm import tqdm

# === Filsti til tweets data ===
input_path = r"C:\Users\Kasper Hassing\Desktop\Speciale_KryptoSentiment\data\twitter_posts\bitcoin_tweets_cleaned_08.csv"
output_path = r"C:\Users\Kasper Hassing\Desktop\Speciale_KryptoSentiment\data\twitter_posts\bitcoin_tweets_cleaned_09.csv"

# === IndlÃ¦s data ===
print("ğŸ“¥ IndlÃ¦ser tweets...")
df = pd.read_csv(input_path)

# === Rensefunktion for links og hashtags ===
def clean_tweet(text):
    # Fjern links (http, https, www)
    text = re.sub(r'http\S+|www\.\S+', '', text)

    # Fjern hashtags, hvis de stÃ¥r til sidst som en liste
    text = re.sub(r'(\s?#\w+)+$', '', text)

    # Behold hashtags i sÃ¦tninger (ikke fjernet ovenfor)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# === Progress bar til rensning ===
print("\nğŸ§¼ Rensning af tweets (fjerner links og ubrugelige hashtags)...")
tqdm.pandas(desc="Rensning")
df['text_cleaned'] = df['text'].progress_apply(clean_tweet)

# === Statistik fÃ¸r og efter rensning ===
print("\nğŸ“Š Statistik fÃ¸r og efter:")
print(f"Total antal tweets: {len(df):,}")
print(f"Tweets med links (fÃ¸r rensning): {(df['text'].str.contains('http')).sum():,}")
print(f"Tweets med links (efter rensning): {(df['text_cleaned'].str.contains('http')).sum():,}")

# === Gem det rensede datasÃ¦t ===
print(f"\nğŸ’¾ Gemmer det rensede datasÃ¦t til:\n{output_path}")
df.to_csv(output_path, index=False)

print("\nâœ… Rensning fuldfÃ¸rt!")
```

```python
import pandas as pd

# === Filsti til tweets data ===
input_path = r"C:\Users\Kasper Hassing\Desktop\Speciale_KryptoSentiment\data\twitter_posts\bitcoin_tweets_cleaned_11.csv"
output_path = r"C:\Users\Kasper Hassing\Desktop\Speciale_KryptoSentiment\data\twitter_posts\bitcoin_tweets_cleaned_12.csv"

# === LÃ¦s data ===
print("ğŸ“¥ IndlÃ¦ser tweets...")
try:
    # LÃ¦s data i batches for at undgÃ¥ hukommelsesproblemer
    chunk_size = 500000
    filtered_tweets = []

    # Stream data for at spare hukommelse
    for chunk in pd.read_csv(input_path, chunksize=chunk_size, usecols=['username', 'datetime', 'text_clean']):
        # Beregn antal tweets pr. bruger i chunket
        tweet_counts = chunk['username'].value_counts()

        # Identificer brugere med â‰¤ 500 tweets
        valid_users = tweet_counts[tweet_counts <= 500].index

        # Filtrer chunket baseret pÃ¥ gyldige brugere
        filtered_chunk = chunk[chunk['username'].isin(valid_users)]
        filtered_tweets.append(filtered_chunk)

        print(f"âœ… Chunk filtreret: {len(filtered_chunk):,} tweets")

    # Kombiner alle de filtrerede chunks
    final_df = pd.concat(filtered_tweets)

    # Gem det filtrerede datasÃ¦t
    final_df.to_csv(output_path, index=False)
    print(f"\nğŸ’¾ Renset datasÃ¦t gemt som: {output_path}")
    print(f"Antal tweets efter filtrering: {len(final_df):,}")

except FileNotFoundError:
    print(f"âŒ Filen blev ikke fundet: {input_path}")
except Exception as e:
    print(f"ğŸš¨ Der opstod en fejl: {e}")
```
